```python
import pyvista as pv
import tetgen

meshes = [pv.Sphere(radius=i).triangulate() for i in range(10)]
merged = pv.PolyData()
for mesh in meshes:
    merged += mesh

tet = tetgen.TetGen(merged)
tet.tetrahedralize(order=1, nobisect=False, opt_scheme=0, steinerleft=0, minratio=10, verbose=True)
```
